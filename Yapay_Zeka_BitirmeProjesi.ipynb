# **T5 - Metinden Metne dönüştürücü ile metin özetleme**

```python
#Gerekli kütüphanelerin kurulması
import torch #derin öğrenme modelleri oluştururken kullanmak için 
#aşağıda ki kütüphane metinden metne özet oluşturmak için kuruldu . 
from transformers import T5ForConditionalGeneration,T5Tokenizer 
#Önceden eğitilmiş modellerin yüklenmesi
#özet için modelin yüklenmesi
summary_model = T5ForConditionalGeneration.from_pretrained('t5-base')
#Özet için belirtecin yüklenmesi 
summary_tokenizer = T5Tokenizer.from_pretrained('t5-base')
#özet için aygıtının oluşturulması
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
#eğitilmiş modelde aygıtın çalıştırılması 
summary_model = summary_model.to(device)
```

# Metinden Tekrarlanabilir bir rasgele seçim yapma 

```python
#rastgele sorular oluşturmak ve her seferinde farklı 
#soru oluşturmak için random kütüphanesi eklendi . 
import random 
#metinden ile oluşturulmuş dizimiz üzerinde çalışmak için 
#numpy kütüphanesi kuruldu. 
import numpy as np

#Rastgelelik içeren işlemlerde herkesin aynı sonucu bulabilmesi için kullanıldı.
def set_seed(seed: int): 
    random.seed(seed) #öğenin rastgele seçilmesi
    np.random.seed(seed)#Her çalışmada  dizinden farklı seçimler yapma  
    #Her çalışmadada cekirdeği ayarlar 
    torch.manual_seed(seed) 
    #rasgele sayı oluşturucunun çekirdeğinin zamanla değişmesini sağlamak
    torch.cuda.manual_seed_all(seed)
    
set_seed(42) #başlangıç değerini vererek fonksiyon çağrıldı.
```

# Özetlenmiş Metin Oluşturma ve Veri Seti Eğitimi

```python
#Natural Language Toolkit olan bir doğal dil işleme kütüphanesidir.
#Girilne metin üzerinden çalışabilmek için kuruldu.
import nltk
#gerekli veri setlerinin indirilmesi . 
nltk.download('punkt')
nltk.download('brown')
nltk.download('wordnet')
#WordNet, başka bir NLTK korpus okuyucusudur.Bu yüzden import ettim.
from nltk.corpus import wordnet as wn

#Kütüphanemi import edip yapmak istediğim işlemler için nltk’da bulunan 
#tokenize modülünü kullandım. 
#iki cümleyi birbirinden ayırmak sent_tokenize için kullandım.
from nltk.tokenize import sent_tokenize

def postprocesstext (content):#İşlem sonrası metin 
  final="" #metin sonunda bosluğa ulaşıldığında final 
  for sent in sent_tokenize(content): 
    sent = sent.capitalize() #cümlenin ilk harfini büyük harfle yazma
    #cümle sonununda başka cümleye geçildiğinde büyük harfle başlaması sağlandı.
    final = final +" "+sent 
  return final #final değeri dönderildi.

#METİN ÖZETLEME FONKSİYONU
def summarizer(text,model,tokenizer):
  #String’in başındaki ve sonundaki boşluğu strip() fonksiyonu ile temizliyoruz.
  #Daha sonra alta geçişlere boşluk koyuldu.
  text = text.strip().replace("\n"," ")
  # özet metnin yazdırılması 
  text = "özet: "+text 
  print (text)
  max_len = 512 #maximum metin uzunluğu
  #Her veri farklı uzunluktaolabilmekte 
  #Çoğu algoritma giriş verilerinin sabit uzunlukta olmasını bekler.
  #Bu durumlar için tokenizer’ın padding metodunu kullandım.

#Bu bölümde WordPiece dayalı bir BERT tokenizer’ı oluşturuldu.

  #tokenizer.encode_plus() yalnızca bir değerler listesi yerine 
  #özellikle bir değerler sözlüğü döndürür.
  #tokenizer.encode_plus() dikkat_maskları ve belirteç türü
  #kimlikleri gibi birçok farklı türde bilgi döndürebildiğinden, 
  #her şey bir sözlük biçiminde döndürülür bu yüzden 
  #tokenizer.encode_plus() kullanmayı tercih ettim.

  encoding = tokenizer.encode_plus(text,max_length=max_len,#max uzunluk verildi. 
                              #sabit uzunlukta olma zorunluluğuna false atandı.
                                   pad_to_max_length=False, 
                              #Bir seferde yalnızca 512 belirteci kabul/giriş
                              #olarak alabildiğinden, truncation parametresini 
                              #True olarak belirtmeliyiz.
                                   truncation=True,
                              #return_tensors = "pt",yalnızca belirteç 
                              #oluşturucunun PyTorch 
                              #tensörlerini döndürmesi içindir
                                   return_tensors="pt").to(device)
                                   

  #Maskeli Dil Modelleme boşlukları doldurma görevidir. 
  #attention_mask, dizileri bir araya toplarken kullanılan isteğe bağlı
  #bağımsız değişkendir.Bu argüman, modele hangi belirteçlerin dikkate alınması 
  #ve hangilerinin alınmaması gerektiğini gösterir.
  input_ids, attention_mask = encoding["input_ids"], encoding["attention_mask"]
  #input_ids’ etiketine sahip verileri ve maskelenmiş verileri ayırmalıyız.
  #(‘attention_mask’)

#Text Generation : Metin oluşturma, yeni metin üretme görevidir.
  outs = model.generate(input_ids=input_ids,
                        #Belirteç indekslerini doldururken maskeleme yapılması 
                                  attention_mask=attention_mask,
                        #Grup başına en az tümce bittiğinde aramayı durdurmak 
                                  early_stopping=True, #için true değer atandı.
                        # isteğe bağlı model yapılandırma değeri 3 verildi.
                                  num_beams=3,
                        #her öğe için bağımsız olarak hesaplanan 
                        #döndürülen dizilerin sayısı 1 verildi.
                                  num_return_sequences=1,
                        #bu boyuttaki tüm ngramlar yalnızca 2 kez oluşabilir.
                                  no_repeat_ngram_size=2,
                        #Oluşturulan dizinin minimum uzunluğu.
                                  min_length = 75,
                        #Oluşturulan dizinin sahip olabileceği maksimum uzunluk. 
                                  max_length=300)

  # oluşturulan tokenları yani belirteçleri çalıştırabilmek için kodu çözüyoruz. 
  dec = [tokenizer.decode(ids,skip_special_tokens=True) for ids in outs]
  summary = dec[0] #çözülen kod dizisinde ilk elamnı alıyoruz.
  summary = postprocesstext(summary) #oluşturulan veriyi çekiyoruz.
  summary= summary.strip() #dizenin başındaki ve sonundaki boşlukları kaldırdım.

  return summary # özet metin değeri dönderildi.

#fonksiyona girilen metin verildi değeri , özeti, özet tokenları gönderildi
summarized_text = summarizer(text,summary_model,summary_tokenizer)
#özet metin alındı.

#Esas metinin ekrana bastırılması .
print ("\nOrijinal Metin >>")
for wrp in wrap(text, 150): 
  print (wrp)
print ("\n")
#Özet metnin ekrana bastırılması .
print ("Ozetlenmis metin >>")
for wrp in wrap(summarized_text, 150):
  print (wrp)
print ("\n")
```

# **Anahtar Sözcükleri Bulma**

```python
import nltk #Doğal Dil işleme kütphanesi
#anahtar kelime bulmak için gereksiz edatların dekarte edilmesi. 
nltk.download('stopwords') #için gerekli kütphane
from nltk.corpus import stopwords #kütüphanenin import edilmesi
import string #metin verilerini kullanabilmek için 
import pke #uçtan uca bir anahtar sözcük çıkarma hattı sağlar.
import traceback #yığın izleme için kullanıldı. 

def get_nouns_multipartite(content): #isimleri alma anahtar kelimeleri alma 
    out=[]
    try:
        extractor = pke.unsupervised.MultipartiteRank() #sıralama yapar
        extractor.load_document(input=content,language='en') #dil secici 
        # noktalama işaretlerini yok sayar. 
        pos = {'PROPN','NOUN'} #özel isim ve karşıtı 
        stoplist = list(string.punctuation) #noktalama işaretlerini listele 
        #edatların alınması 
        stoplist += ['-lrb-', '-rrb-', '-lcb-', '-rcb-', '-lsb-', '-rsb-']
        #word kütüphanesinin desteklediği dilin seçimi 
        stoplist += stopwords.words('english')
        # extractor.candidate_selection(pos=pos, stoplist=stoplist)
        extractor.candidate_selection(pos=pos) #çalıştırılması
        # Rastgele yürütme kullanarak anahtar secer ve Eşik parametresi ayarlar
        extractor.candidate_weighting(alpha=1.1, #eş sözcükleri alan parametre 
                                      #kümeleme için minimum benzerlik
                                      threshold=0.75,
                                      #bağlanti yöntemi, varsayılan seçildi.
                                      method='average')
    #bağlantı yönteminen yüksek  eşik parametresi (15) puan alan adayı alma
        keyphrases = extractor.get_n_best(n=15) 
        
 #bağlantı yönteminen yüksek  eşik parametresi (15) puan alan 
 #adayı listeden yazdırma 
        for val in keyphrases:
            out.append(val[0]) 
    except:
        out = []
        traceback.print_exc() #yığından cıkılır 

    return out #cıkış dönülür
```

```python
!pip install flashtext #flash kütüphanesi kullanabilmek için 
!pip install --upgrade scipy networkx #version hatası almamak için güncellendi.
from flashtext import KeywordProcessor #web api ile işlem yapabilmek için 


def get_keywords(originaltext,summarytext):
  keywords = get_nouns_multipartite(originaltext)
  print ("keywords unsummarized: ",keywords)
  keyword_processor = KeywordProcessor()
  for keyword in keywords:
    keyword_processor.add_keyword(keyword)

  keywords_found = keyword_processor.extract_keywords(summarytext)
  keywords_found = list(set(keywords_found))
  print ("keywords_found in summarized: ",keywords_found)

  important_keywords =[]
  for keyword in keywords:
    if keyword in keywords_found:
      important_keywords.append(keyword)

  return important_keywords[:4]


imp_keywords = get_keywords(text,summarized_text)
print (imp_keywords)
```

```python
def get_question(context,answer,model,tokenizer): #soru oluşturma fonksiyonu
  text = "context: {} answer: {}".format(context,answer) #soru-cevap
  encoding = tokenizer.encode_plus(text,max_length=384, pad_to_max_length=False,truncation=True, return_tensors="pt").to(device)
  #Maskeli Dil Modelleme boşlukları doldurma görevidir. 
  #attention_mask, dizileri bir araya toplarken kullanılan isteğe bağlı
  #bağımsız değişkendir.Bu argüman, modele hangi belirteçlerin dikkate alınması 
  #ve hangilerinin alınmaması gerektiğini gösterir.
  input_ids, attention_mask = encoding["input_ids"], encoding["attention_mask"]
   #tokenizer.encode_plus() yalnızca bir değerler listesi yerine 
  #özellikle bir değerler sözlüğü döndürür.
  #tokenizer.encode_plus() dikkat_maskları ve belirteç türü
  #kimlikleri gibi birçok farklı türde bilgi döndürebildiğinden, 
  #her şey bir sözlük biçiminde döndürülür bu yüzden 
  #tokenizer.encode_plus() kullanmayı tercih ettim.
  outs = model.generate(input_ids=input_ids,
                        #Belirteç indekslerini doldururken maskeleme yapılması 
                                  attention_mask=attention_mask,
                         #Grup başına en az tümce bittiğinde aramayı durdurmak 
                                  early_stopping=True,
                        # isteğe bağlı model yapılandırma değeri 5 verildi.
                                  num_beams=5,
                        #her öğe için bağımsız olarak hesaplanan 
                        #döndürülen dizilerin sayısı 1 verildi.
                                  num_return_sequences=1,
                           #Oluşturulan dizinin minimum uzunluğu.
                                  no_repeat_ngram_size=2,
                          #Oluşturulan dizinin sahip olabileceği maksimum uzunluk.
                                  max_length=72)

# oluşturulan tokenları yani belirteçleri çalıştırabilmek için kodu çözüyoruz. 
  dec = [tokenizer.decode(ids,skip_special_tokens=True) for ids in outs]

#oluşturulan sorunun alınması ve kullanılması 
  Question = dec[0].replace("question:","")
  Question= Question.strip()
  return Question

#sorunun hangi özet cümleden alındığını anlamak için 
#alındığı kısımdaki özet cümlenin getirilmesi. 
for wrp in wrap(summarized_text, 150):
  print (wrp)
print ("\n")
#özet metindeki sorunun oluşumuna yardım eden anahtar sözcüğün cümlede bulunması 
#belirgin olması için anahtar sözcüğün kalın fontta gösterilmesi.
for answer in imp_keywords:
  ques = get_question(summarized_text,answer,question_model,question_tokenizer)
  print (ques) #sorunun bastırılması
  print (answer.capitalize()) #sorudaki anahtar kelime
  print ("\n")
```

#**1.1-Metin Benzerliklerini Bulma**

```python
# metin benzerliğini bulmaK için import edildi.
from similarity.normalized_levenshtein import NormalizedLevenshtein 
normalized_levenshtein = NormalizedLevenshtein()

def filter_same_sense_words(original,wordlist):#aynı anlamlı kelimlerin filtrelenmesi fonksiyonu
  filtered_words=[] #filtrelenmis metin dizisi
  base_sense =original.split('|')[1] #temel anlam 
  print (base_sense) #ekrana basılması
  #tüm kelimelerin filtrelenmesi
  for eachword in wordlist: 
    if eachword[0].split('|')[1] == base_sense:
      filtered_words.append(eachword[0].split('|')[0].replace("_", " ").title().strip())
  return filtered_words
#en iyi anahtar kelimelerin secilmesi 
def get_highest_similarity_score(wordlist,wrd):
  score=[]
  for each in wordlist:
    score.append(normalized_levenshtein.similarity(each.lower(),wrd.lower()))
  return max(score)
#metin benzerliklerinin bulunması 
def sense2vec_get_words(word,s2v,topn,question):
    output = []#cıkış boşaltıldı
    print ("kelime ",word) #kelimenin ekrana basılması
    try:
      sense = s2v.get_best_sense(word, senses= ["NOUN", "PERSON","PRODUCT","LOC","ORG","EVENT","NORP","WORK OF ART","FAC","GPE","NUM","FACILITY"])
      #isim,ürün,konum,organizasyon... bilgilerin alınması 
     #sonuçarın görüntülnmesi 
      most_similar = s2v.most_similar(sense, n=topn) #maximum benzerlik sayısına(eşik değerine)sahip değer atandı 
      output = filter_same_sense_words(sense,most_similar)#benzer kelimler filtrelendi ve çıkışa aktarıldı.
      print ("Benzer ",output) #cıkış bastırıldı
    except:
      output =[] #cıkış boşaltıldı 
    threshold = 0.6 #eşik değeri belirtldi 
    final=[word] #kelimenin liste değeri alındı
    checklist =question.split() #kontrol listesine soru atandı 
    for x in output: #cıkış dongüsüyle benzer kelimeler listeye eklendi.
      if get_highest_similarity_score(final,x)<threshold and x not in final and x not in checklist:
        final.append(x) #final kelime listeye eklendi
    return final[1:]

def mmr(doc_embedding, word_embeddings, words, top_n, lambda_param):

   # Kelimeler içindeki ve kelimeler ile belge arasındaki benzerliği ayıklama
    word_doc_similarity = cosine_similarity(word_embeddings, doc_embedding)
    word_similarity = cosine_similarity(word_embeddings)

  #Var olan anahtar kelimelerin alınması ve en iyi anahtar kelimenin/ anahtar cümlenin seçilmesi
    keywords_idx = [np.argmax(word_doc_similarity)]
    candidates_idx = [i for i in range(len(words)) if i != keywords_idx[0]]

    for _ in range(top_n - 1):
       # Kelime benzerlikleri ayıklama ve
       # seçilen anahtar kelimeler / kelime öbekleri arasında seçim yapma
        candidate_similarities = word_doc_similarity[candidates_idx, :]
        target_similarities = np.max(word_similarity[candidates_idx][:, keywords_idx], axis=1)

      #mmr değerinin ehsaplanması 
        mmr = (lambda_param) * candidate_similarities - (1-lambda_param) * target_similarities.reshape(-1, 1)
        mmr_idx = candidates_idx[np.argmax(mmr)]

     # Anahtar kelimeleri ve adaylarını güncelleme
        keywords_idx.append(mmr_idx)
        candidates_idx.remove(mmr_idx)

    return [words[idx] for idx in keywords_idx] #anahtar kelimelerin ve ifadelerin dönülmesi 
```

```python
#collections, farklı nesneleri depolamak ve içerilen nesnelere erişmenin ve bunlar üzerinde yineleme yapmak için eklendi. 
# Yinelemeli öğelerin sayısını tutmak OrderedDict import edildi . 
from collections import OrderedDict
# geçerli iki vektör arasındaki kosinüs benzerliğinin bulunması için import edildi .   
from sklearn.metrics.pairwise import cosine_similarity
import nltk #doğal dil işleme kütüphanesi
nltk.download('omw-1.4') #öğeler için indirildi.

def get_distractors_wordnet(word): #dikkat dağıtıcı oluşturma fonksiyonu
    distractors=[]
    try:
      syn = wn.synsets(word,'n')[0] #ilk değer 
      
      word= word.lower() #kelime kücük harflere dönüştürüldü. 
      orig_word = word
      if len(word.split())>0: #kelimenin uzunluğu 0 dan büyükse 
          word = word.replace(" ","_") #görülen şekilde kelimenin atanması
      hypernym = syn.hypernyms() #benzer sözcükler türetildi 
      if len(hypernym) == 0: #klime uzunluğu sıfır ise
          return distractors # 
      for item in hypernym[0].hyponyms(): #benzer sözcük türetme 
          name = item.lemmas()[0].name() #sözcüğün eklenmesi ve atanması 
          #print ("name ",name, " word",orig_word)
          if name == orig_word: #orjinal kelimeye eşitse 
              continue #geri dön
          name = name.replace("_"," ") #kelimenin alınması 
          name = " ".join(w.capitalize() for w in name.split())#konteynıra ekleme 
          if name is not None and name not in distractors: #yanlış şıkkın
              distractors.append(name)#eklemesi
    except:
      print ("Wordnet Kütüphanesi dağıtıcıları(yanlış şıkları) bulunamadı")
    return distractors #dikkat dağıtıların dönülmesi 

def get_distractors (word,origsentence,sense2vecmodel,sentencemodel,top_n,lambdaval): #dikkat dağıtıcı fonksiyonu
  distractors = sense2vec_get_words(word,sense2vecmodel,top_n,origsentence) #dikkat dağıtıcının alınması
  print ("DİKKAT DAĞITICI: ",distractors) #ekrana bastırılması 
  if len(distractors) ==0: #uzunluğu 0 ise 
    return distractors #dağıtının dönülmesi
  distractors_new = [word.capitalize()] #yeni dikkat dağıtı
  distractors_new.extend(distractors) #yeni dikkat dağıtıcıın genişletilmesi
  # print ("distractors_new .. ",distractors_new)

  #cümlelerin konteynıra gömülmesi
  embedding_sentence = origsentence+ " "+word.capitalize() 
  # embedding_sentence = word
  #anahtar cümlelerin konteynıra gömülmesi
  keyword_embedding = sentencemodel.encode([embedding_sentence])
  distractor_embeddings = sentencemodel.encode(distractors_new)#yeni konteynıra eklenip değerin alınması 

  # filtered_keywords = mmr(keyword_embedding, distractor_embeddings,distractors,4,0.7)
  max_keywords = min(len(distractors_new),5) #maximum değere sahip anahtar kelime
  filtered_keywords = mmr(keyword_embedding, distractor_embeddings,distractors_new,max_keywords,lambdaval)#anahtar kelimelerin filtrelenmesi
  # filtered_keywords = filtered_keywords[1:]
  final = [word.capitalize()] #fonksiyonun çağrılması ve listedeki değerinin finale eklenmesi 
  for wrd in filtered_keywords:#anahtar kelimlerin filtrelenmesi
    if wrd.lower() !=word.lower(): #kelimeler kücük harf değilse
      final.append(wrd.capitalize()) #final listesine ekleme 
  final = final[1:] #final değerin yenilenmesi
  return final

#deneme 
sent = "What cryptocurrency did Musk rarely tweet about?" #soru
keyword = "Bitcoin"#anahtar kelime 

# sent = "What did Musk say he was working with to improve system transaction efficiency?"
# keyword= "Dogecoin"


# sent = "What company did Musk say would not accept bitcoin payments?"
# keyword= "Tesla"


# sent = "What has Musk often tweeted in support of?"
# keyword = "Cryptocurrency"

print (get_distractors(keyword,sent,s2v,sentence_transformer_model,40,0.2)) # anahtar keilmenin  benzerlerinin ekrana basılması
```
